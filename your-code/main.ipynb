### Part 0 - Preliminary Step: Remove outliers in product prices.

For this purpose, we can use the IQR (Interquartile Range) method. Products priced below the first quartile minus 1.5 times the IQR or above the third quartile plus 1.5 times the IQR will be considered outliers and removed from the dataset. The next steps will be done with the dataframe without outliers.

**Hint**: You can check the Descriptive Stats lesson for a reminder on Inter Quartile Range (IQR).

```python
# your code here
df = pd.read_csv('amz_uk_price_prediction_dataset.csv.zip', compression='zip')
df.head()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Determine the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

# Display the DataFrame without outliers
print("\nDataFrame without outliers:")
print(df_no_outliers.head())
```

### Part 1 - Violin Plots:

Use a violin plot to visualize the distribution of price across different product categories. Filter out the top 20 categories based on count for better visualization.

Which product category tends to have the highest median price? Don't filter here by top categories.

```python
# your code here
top_20_categories = df['category'].value_counts().nlargest(20).index
df_filtered = df[df['category'].isin(top_20_categories)]
df_filtered.head()
```

```python
# Create a violin plot
g = sns.FacetGrid(df_filtered, col='category', col_wrap=5, height=4, sharey=False)

# Map the violin plot to each facet
g.map(sns.violinplot, 'price', color='skyblue')

# Add titles and labels
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Price", "Category")
g.add_legend()

# Display the plot
plt.show()
#the plots look wierd... Why would we choose violin plot here?  Doesnt look sutable at all. Even if plotting separately for each category.
```

### Part 2 - Bar Charts:

Create a bar chart comparing the average price of products for the top 10 product categories (based on count).
Which product category commands the highest average price? Don't filter here by top categories.

```python
# your code here
avg_price_by_category = df.groupby('category')['price'].mean()
```

```python
top_10_categories = df['category'].value_counts().nlargest(10).index
```

```python
avg_price_top_10 = avg_price_by_category[top_10_categories]
```

```python
plt.figure(figsize=(12, 6))
plt.bar(avg_price_top_10.index, avg_price_top_10.values)

# Add titles and labels
plt.title('Average Price of Products by Top 10 Categories')
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

### Part 3 - Box Plots:

Visualize the distribution of product ratings based on their category using side-by-side box plots. Filter out the top 10 categories based on count for better visualization.
Which category tends to receive the highest median rating from customers? Don't filter here by top categories.

```python
# your code here
top_10_categories = df['category'].value_counts().nlargest(10).index
df_filtered = df[df['category'].isin(top_10_categories)]
plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='stars', data=df_filtered)
plt.xticks(rotation=45, ha='right')
plt.title('Distribution of Product Ratings by Category (Top 10)')
plt.show()
```
